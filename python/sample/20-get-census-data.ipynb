```python
import pandas as pd
import numpy as np

acsLoc = '/Volumes/Echo/GIS/projects/naas/data/input/census/tables/acs5yr/2012/'
cenLoc = '/Volumes/Echo/GIS/projects/naas/data/input/census/tables/census/2000/'

pov00 = pd.read_csv(cenLoc+'tract_st36_pov.csv')
pop00 = pd.read_csv(cenLoc+'tract_st36_pop.csv')
mhi00 = pd.read_csv(cenLoc+'tract_st36_mhi.csv')
pov10 = pd.read_csv(acsLoc+'tract_st36_pov.csv')
pop10 = pd.read_csv(acsLoc+'tract_st36_pop.csv')
mhi10 = pd.read_csv(acsLoc+'tract_st36_mhi.csv')

tList = [pov00,pop00,mhi00,pov10,pop10,mhi10]

for df in tList:
    df['state'] = df['state'].astype(str)
    df['county'] = df['county'].astype(str)
    df['tract'] = df['tract'].astype(str)
    
    df['county'] = df['county'].apply(lambda x: '{0:0>3}'.format(x)) #add leaving zeros
    df['tract'] = df['tract'].apply(lambda x: '{0:0>6}'.format(x))
    df = df.rename(columns=lambda x: x.replace('_', ''))
    
pov00['geoid'] = pov00['state'] + pov00['county'] + pov00['tract']
pop00['geoid'] = pop00['state'] + pop00['county'] + pop00['tract']
mhi00['geoid'] = mhi00['state'] + mhi00['county'] + mhi00['tract']
pov10['geoid'] = pov10['state'] + pov10['county'] + pov10['tract']
pop10['geoid'] = pop10['state'] + pop10['county'] + pop10['tract']
mhi10['geoid'] = mhi10['state'] + mhi10['county'] + mhi10['tract']

pov00 = pov00.drop(['state','county','tract'], axis=1)
pop00 = pop00.drop(['state','county','tract'], axis=1)
mhi00 = mhi00.drop(['state','county','tract'], axis=1)
pov10 = pov10.drop(['state','county','tract'], axis=1)
pop10 = pop10.drop(['state','county','tract'], axis=1)
mhi10 = mhi10.drop(['state','county','tract'], axis=1)

pov00.to_csv(cenLoc+'tract_st36_pov_cln.csv',index=False)
pop00.to_csv(cenLoc+'tract_st36_pop_cln.csv',index=False)
mhi00.to_csv(cenLoc+'tract_st36_mhi_cln.csv',index=False)
pov10.to_csv(acsLoc+'tract_st36_pov_cln.csv',index=False)
pop10.to_csv(acsLoc+'tract_st36_pop_cln.csv',index=False)
mhi10.to_csv(acsLoc+'tract_st36_mhi_cln.csv',index=False)

pov00.head(10)
```
